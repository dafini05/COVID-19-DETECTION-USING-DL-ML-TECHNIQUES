```python
import cv2
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
def loadImages(path, urls, target):
  images = []
  labels = []
  for i in range(len(urls)):
    img_path = path + "/" + urls[i]
    img = cv2.imread(img_path)
    img = img / 255.0
    #print(img_path)
    # if we want to resize the images
    img = cv2.resize(img, (100, 100))
    images.append(img)
    labels.append(target)
  images = np.asarray(images)
  return images, labels
```

```python
covid_path = "/content/COVID19_Pneumonia_Normal_Chest_Xray_PA_Dataset/covid"
covidUrl = os.listdir(covid_path)
covidImages, covidTargets = loadImages(covid_path, covidUrl, 1)
```

```python
normal_path = "/content/COVID19_Pneumonia_Normal_Chest_Xray_PA_Dataset/normal"
normal_urls = os.listdir(normal_path)
normalImages, normalTargets = loadImages(normal_path, normal_urls, 0)
```

```python
data = np.r_[covidImages, normalImages]
```

```python
 targets = np.r_[covidTargets, normalTargets]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, targets, test_size=0.30)
```

```python
import tensorflow as tf
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x_train = np.array(x_train).reshape(x_train.shape[0], -1)
x_test = np.array(x_test).reshape(x_test.shape[0], -1)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
k_values = list(range(1, 21))
accuracies = []
```

```python
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracies.append(accuracy_score(y_test, y_pred))
```

```python
best_k = k_values[accuracies.index(max(accuracies))]
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Best k: {best_k}')
print(f'Accuracy: {accuracy}')
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'Confusion Matrix:\n{conf_matrix}')
class_report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{class_report}')
```
